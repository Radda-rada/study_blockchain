```python
import hashlib

def find_nonce(input_text, num_zeros):
    nonce = 0
    while True:
        data = input_text + str(nonce)
        hash_result = hashlib.sha256(data.encode()).hexdigest()
        if hash_result[:num_zeros] == '0' * num_zeros:
            return nonce
        nonce += 1

input_text = "hello how r u"

# Поиск Nonce для 1 нуля
nonce_1_zero = find_nonce(input_text, 1)
print(nonce_1_zero)

# Поиск Nonce для 2 нулей
nonce_2_zeros = find_nonce(input_text, 2)
print(nonce_2_zeros)

# Поиск Nonce для 3 нулей
nonce_3_zeros = find_nonce(input_text, 3)
print(nonce_3_zeros)
```

```python
import rsa
import hashlib

message = "passion for crypto"

# Генерация ключей
pub_key, priv_key = rsa.newkeys(2048)

# Хэширование сообщения по алгоритму SHA256
hashed_message = hashlib.sha256(message.encode()).digest()

# Выработка подписи
signature = rsa.sign(hashed_message, priv_key, 'SHA-256')

# Преобразование подписи в шестнадцатеричный формат
signature_hex = signature.hex()

print("Подпись:", signature_hex)
```

```python
from ecdsa import SigningKey

# Генерация приватного ключа (ключа подписи)
sk = SigningKey.generate()

# Получение публичного ключа (ключа верификации) из приватного ключа
vk = sk.verifying_key

Mes = input("Введите сообщение: ")
Mes = Mes.encode()

# Подпись сообщения с использованием приватного ключа
signature = sk.sign(Mes).hex()

# Верификация подписи с использованием публичного ключа
verify = vk.verify(bytes.fromhex(signature), Mes)

print("Публичный ключ (ключ верификации):", vk.to_string().hex())
print("Подпись:", signature)
print("Верификация подписи:", verify)
```

```python
import random

def fiat_shamir():
    # Генерация случайного числа
    s = random.randint(1, 1000)
    
    # Вычисление публичного ключа
    v = s ** 2
    
    # Вычисление закрытого ключа
    sk = s
    
    return sk, v

# Генерация пары ключей по протоколу Фиата-Шамира
sk, v = fiat_shamir()

# Получение первых 4 найденных символов ключей
sk = str(sk)[:4]
v = str(v)[:4]

print("Закрытый ключ:", sk)
print("Публичный ключ:", v)
```

```python
def fiat_shamir():
    # Ввод двух чисел с клавиатуры
    a = int(input("Введите первое число: "))
    b = int(input("Введите второе число: "))
    
    # Вычисление публичного ключа
    v = a * b
    
    # Вычисление закрытого ключа
    sk = a + b
    
    return sk, v

# Генерация пары ключей по протоколу Фиата-Шамира
sk, v = fiat_shamir()

# Получение первых 4 найденных символов ключей
sk = str(sk)[:4]
v = str(v)[:4]

print("Закрытый ключ:", sk)
print("Публичный ключ:", v)
```

```python
import math

p = int(input())
q = int(input())

n = p*q
def find_V(n):
    V = []
    for i in range(1, n):
        x = i ** 2 % n
        if math.gcd(x, n) == 1:
            V.append(x)
    return V

V = find_V(n)

def find_S (v, n):
    S = []
    for i in v:
        j=1
        while (1):
            if (i*j)%n == 1:
                if math.sqrt(j).is_integer():
                    S.append(math.sqrt(j))
                else:
                    while (1):
                        j+=n
                        if math.sqrt(j).is_integer():
                            S.append(math.sqrt(j)%n)
                            break
                break
            else: j+=1
    return S
S = find_S(V, n)

PubK = V[:4]
PrK = S[:4]

print("Public key:", PubK)
print("Private key:", PrK)
```
